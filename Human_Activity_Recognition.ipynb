### Creating X and Y 

```python
xtrain = train.drop('Activity', axis=1)
ytrain =train.loc[: ,'Activity']

xtest =test.drop('Activity' ,axis=1)
ytest =test.loc[:,'Activity']
```

### Modeling Phase

#### Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
model = GaussianNB()
```

```python
model.fit(xtrain,ytrain)
```

```python
pred= model.predict(xtest)
pred[:100]
```

### Confusion Matrix

```python
plt.figure(figsize=(6,6))
cm = confusion_matrix(pred, ytest)

sns.heatmap(cm, square=True , annot=True ,fmt='d')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.plot()
```

```python
from sklearn import metrics
print(metrics.classification_report(pred, ytest, labels=list(train['Activity'].unique())))
```

### Accuracy

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the model:" , accuracy_score(ytest,pred))
```
